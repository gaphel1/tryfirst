```python
import numpy as np
import pandas as pd
import datetime
import matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
data_nut = pd.read_excel('nurtr.xlsx')
```

```python
data_nut.info()
```

```python
data_nut.drop_duplicates(inplace=True)
#removing duplicates for cleaning data
```

```python
data_nut.tail(11)
```

```python
data_nut.user_id
```

```python
data_nut['created_at'].max()
```

```python
import datetime as dt
sd = dt.datetime(2019,2,20)

data_nut['hist']=sd - data_nut['created_at']
data_nut['hist'].astype('timedelta64[D]')
data_nut['hist']=data_nut['hist'] / np.timedelta64(1, 'D')
data_nut.head()
# forrmla to calculate the frequency
data_nut=data_nut[data_nut['hist'] < 150]
data_nut.info()
```

```python
rfTable = data_nut.groupby('user_id').agg({'hist': lambda x:x.min(), # Recency
                                        'user_id': lambda x: len(x),# Frequency
                                       
                                         }) 
```

```python
rfTable.rename(columns={'hist': 'recency',
                         'user_id': 'frequency', 
                        }, inplace=True
                         )
```
